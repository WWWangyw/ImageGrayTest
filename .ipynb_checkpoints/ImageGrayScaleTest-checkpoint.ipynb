```python
import cv2
from matplotlib import pyplot as plt
```

```python
img=cv2.imread('image2.jpg',1)
```

```python
plt.hist(img.reshape([-1]),256,[0,256]);
plt.show()
```

```python
color=('b','g','r')
```

```python
for i,col in enumerate(color):
    histr=cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.show
```
